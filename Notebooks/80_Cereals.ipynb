### Plotting Overall distribution of data

```python
plt.figure(figsize=(8,6))
plt.title('Data Distribution')
sns.histplot(data=df)
```

### Cereals with Highest Sugar Content

```python
top_sug = df.loc[df['sugars'] >= 13]
sort_top_sug=top_sug.sort_values(by='sugars',ascending=False)

plt.figure(figsize=(5,3))
plt.title('Cereal with highest Sugar Content')
plt.xticks(rotation=90)
sns.barplot(x=sort_top_sug['name'],y=sort_top_sug['sugars'], data=sort_top_sug)
```

##### The top 5 Cereals Brands contains highest sugar:
    1) Golden Crisp
    2) Smacks
    3) Apple Jacks
    4) Post Nat. Raisin Bran
    5) Total Raisin Bran

### Cereals with Lowest Sugar Content

```python
sort_data=df.sort_values(by='sugars',ascending=False)
top_sug = df.loc[df['sugars'] <= 2]
sort_top_sug=top_sug.sort_values(by='sugars',ascending=True)

plt.figure(figsize=(5,3))
plt.title('Cereal with highest Sugar Content')
plt.xticks(rotation=90)
sns.barplot(x=sort_top_sug['name'],y=sort_data['sugars'], data=sort_data)
```

##### The top 5 Cereals Brands contains Lowest/Zero sugar:
    1) Quacker Oatmeal
    2) All-Bran with Extra Fiber
    3) Cream of Wheat (Quick)
    4) Puffed Rice
    5) Puffed Wheat

### Cereals with Highest Protein Content

```python
sort_pr = df.loc[df['protein'] >=4]
sort_prot = sort_pr.sort_values(by='protein', ascending=False)

plt.figure(figsize=(5,3))
plt.title('Cereals with High protein')
plt.xticks(rotation=90)
sns.barplot(x=sort_prot['name'], y=sort_prot['protein'], data=sort_prot)
```

##### Top 5 Cereals with highest protein Content:
    1) Cheerios
    2) Special K
    3) Quaker Oatmeal
    4) 100% Bran
    5) All-Bran
